```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('E:\Practice Data\Customer Churn.csv')
df.head(30)
```

```python
df.info()
```

#replacing blanks with '0' as tenure is '0' and no total charges are recorded.

```python
df["TotalCharges"] = df["TotalCharges"].replace(' ','0')
df["TotalCharges"] = df["TotalCharges"].astype('float')
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

#Converted 0 and 1 values of senior citizen to yes/no to make it easier to understand.

```python
plt.figure(figsize = (3,4))
ax = sns.countplot(x = 'Churn',data = df)
ax.bar_label (ax.containers[0]) 
plt.title("Count of Churned Customers",fontsize = 10)
plt.show()
```

```python
plt.figure(figsize = (3,4))
# Group the DataFrame by 'Churn' and count occurrences
gb = df.groupby("Churn").agg({'Churn':"count"}) 
# Create a pie chart using the grouped data
plt.pie(gb['Churn'],labels = gb.index,autopct = "%1.2f%%")
# Add a title in the pie Chart
plt.title("Percentage of Churned Customers", fontsize = 10)
# Display the pie chart
plt.show()
```

```python
plt.figure(figsize = (3,5))
ax = sns.countplot(x = "SeniorCitizen",data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Senior Citizen")
plt.show()
```

```python
plt.figure(figsize=(6, 5))  # Adjust the figure size if needed
ax = sns.countplot(x="SeniorCitizen", data=df, hue="Churn")  # Grouped by 'Churn'
ax.bar_label(ax.containers[0])  # Add labels for the first container
ax.bar_label(ax.containers[1])  # Add labels for the second container if present
plt.title("Count of Senior Citizen Customers by Churn Status")
plt.show()
```

```python
# Calculate the counts for each combination of SeniorCitizen and Churn
df_grouped = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()

# Create a stacked bar chart
ax = df_grouped.plot(kind='bar', stacked=True, figsize=(6, 5), color=['#1f77b4', '#ff7f0e'])

# Add data labels on top of each bar
for p in ax.patches:
    width = p.get_width()    # Width of bar (should be constant, as this is vertical bar chart)
    height = p.get_height()  # Height of the bar
    x, y = p.get_xy()        # Bottom left corner of the bar
    if height > 0:
        ax.text(x + width / 2, y + height / 2, f'{int(height)}', ha='center', va='center', color='white', fontsize=10)

# Add title and labels
plt.title("Count of Senior Citizen Customers by Churn Status (Stacked)")
plt.xlabel("Senior Citizen")
plt.ylabel("Count")

# Display the chart
plt.show()
```

```python
# Calculate the counts for each combination of SeniorCitizen and Churn
df_grouped = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()

# Calculate the row-wise sum to get the total count for each SeniorCitizen category
totals = df_grouped.sum(axis=1)

# Create a stacked bar chart
ax = df_grouped.plot(kind='bar', stacked=True, figsize=(6, 5), color=['#1f77b4', '#ff7f0e'])

# Add percentage labels on top of each bar
for p in ax.patches:
    width = p.get_width()    # Width of the bar
    height = p.get_height()  # Height of the bar
    x, y = p.get_xy()        # Bottom left corner of the bar
    if height > 0:
        # Calculate the percentage
        total = totals[int(x + width / 2)]
        percentage = 100 * height / total
        ax.text(x + width / 2, y + height / 2, f'{percentage:.2f}%', ha='center', va='center', color='white', fontsize=10)

# Add title and labels
plt.title("Percentage of Senior Citizen Customers by Churn Status (Stacked)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")

# Display the chart
plt.show()
```

```python


# Assuming you have a DataFrame called df with the columns you mentioned
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 
           'StreamingTV', 'StreamingMovies']

# Define the number of rows and columns for the subplot grid
n_rows = 3
n_cols = 3

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 10))
fig.tight_layout(pad=4.0)

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Loop over the columns and create a count plot for each
for i, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i], hue=df["Churn"])
    axes[i].set_title(f'Count plot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

# Show the plot
plt.show()
```

The chart shows the distribution of various telecom services (e.g., PhoneService, InternetService) with churn data. Most churn occurs among customers with no additional services like OnlineSecurity, #OnlineBackup, or TechSupport. Fiber optic users also exhibit higher churn rates. Overall, customers with fewer or no services tend to churn more often.
